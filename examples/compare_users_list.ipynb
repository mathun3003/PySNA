```python
from pysna import TwitterAPI, TwitterAppAuthHandler

auth = TwitterAppAuthHandler(os.getenv("CONSUMER_KEY"), os.getenv("CONSUMER_SECRET"))
api = TwitterAPI(auth)
```

```python
users = ["Twitter", "YouTube", "BBCWorld"]
num_followers = api.compare_users_list(users, compare="num_followers")
num_followers
```
